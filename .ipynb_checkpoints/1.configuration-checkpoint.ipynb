```python
class Man:
    def __init__(self,name):
        self.name=name
        print("Initialized!")
        return
        
    def hello(self):
        print("Hello " + self.name + " !")
        return
    
    def goodbye(self):
        print("Good Bye " + self.name + " !")
        return
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
m = Man("Kim")
```

```python
m.hello()
m.goodbye()
```

```python
x= np.arange(0,6,0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,label="sin")
plt.plot(x,y2,linestyle="--",label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
